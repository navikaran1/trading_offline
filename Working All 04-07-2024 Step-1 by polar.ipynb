```python
import polars as pl
import os

# File locations
df1_file = r"C:\Users\HP\Desktop\Share market all Working Tool\4_Multi testing\ALL_18_06_2024.csv"
df2_folder = r"D:\scrapy\By polar Step 1\DATA_FOR_TEST"

# Read df1
df1 = pl.read_csv(df1_file)
# Convert the Date column to datetime format
df1 = df1.with_column(pl.col("Date").str.strptime(pl.Date, "%Y-%m-%d"))

# Function to read and process each df2 file
def process_df2_file(file_path):
    df2 = pl.read_csv(file_path)
    # Convert the date column to datetime format
    df2 = df2.with_column(pl.col("date").str.strptime(pl.Date, "%d-%m-%Y"))
    # Join df2 with df1 to lookup the Close price
    joined_df = df2.join(df1.select(["Date", "Symbol", "Close"]),
                         left_on=["symbol", "date"], right_on=["Symbol", "Date"],
                         how="left")
    return joined_df

# List to store results from all df2 files
results = []

# Process each file in df2 folder
for file_name in os.listdir(df2_folder):
    if file_name.endswith('.csv'):
        file_path = os.path.join(df2_folder, file_name)
        result = process_df2_file(file_path)
        results.append(result)

# Concatenate all results
final_result = pl.concat(results)

# Save the final result to a CSV file
final_result.write_csv('final_result.csv')
```

### Finding one day high

```python
import polars as pl
import os

# File locations
df1_file = r"C:\Users\HP\Desktop\Share market all Working Tool\4_Multi testing\ALL_18_06_2024.csv"
df2_folder = r"D:\scrapy\By polar Step 1\DATA_FOR_TEST"

# Create the final data folder in the same directory as df1_file
root_directory = os.path.dirname(df1_file)
final_data_folder = os.path.join(root_directory, "Final Data")
os.makedirs(final_data_folder, exist_ok=True)

print(f"Final data folder created at: {final_data_folder}")

# Read df1
df1 = pl.read_csv(df1_file)
print(f"df1 loaded with {df1.shape[0]} rows and {df1.shape[1]} columns")

# Convert the Date column to datetime format
df1 = df1.with_column(pl.col("Date").str.strptime(pl.Date, "%Y-%m-%d"))
print("Date column converted to datetime")

# Calculate the next day's high price
df1 = df1.with_column(pl.col("High").shift(-1).alias("Next_High"))
print("Next day's high price calculated")

# Function to read and process each df2 file
def process_df2_file(file_path):
    df2 = pl.read_csv(file_path)
    print(f"Processing {file_path} with {df2.shape[0]} rows and {df2.shape[1]} columns")
    # Convert the date column to datetime format
    df2 = df2.with_column(pl.col("date").str.strptime(pl.Date, "%d-%m-%Y"))
    # Join df2 with df1 to lookup the Close price and Next Day High price
    joined_df = df2.join(df1.select(["Date", "Symbol", "Close", "Next_High"]),
                         left_on=["symbol", "date"], right_on=["Symbol", "Date"],
                         how="left")
    return joined_df

# Process each file in df2 folder
for file_name in os.listdir(df2_folder):
    if file_name.endswith('.csv'):
        file_path = os.path.join(df2_folder, file_name)
        result = process_df2_file(file_path)
        # Save the result to a new file in the final data folder
        output_file_path = os.path.join(final_data_folder, f"{os.path.splitext(file_name)[0]}_final_data.csv")
        result.write_csv(output_file_path)
        print(f"Saved result to {output_file_path}")

print("Processing complete. Files saved in 'Final Data' folder.")
```

```python
import polars as pl
import os

# File locations
df1_file = r"C:\Users\HP\Desktop\Share market all Working Tool\4_Multi testing\ALL_18_06_2024.csv"
df2_folder = r"D:\scrapy\By polar Step 1\DATA_FOR_TEST"

# Create the final data folder in the same directory as df1_file
root_directory = os.path.dirname(df1_file)
final_data_folder = os.path.join(root_directory, "Final Data")
os.makedirs(final_data_folder, exist_ok=True)

print(f"Final data folder created at: {final_data_folder}")

# Read df1
df1 = pl.read_csv(df1_file)
print(f"df1 loaded with {df1.shape[0]} rows and {df1.shape[1]} columns")

# Convert the Date column to datetime format
df1 = df1.with_column(pl.col("Date").str.strptime(pl.Date, "%Y-%m-%d"))
print("Date column converted to datetime")

# Calculate the next 7 days' high prices per symbol
for i in range(1, 8):
    df1 = df1.with_column(pl.col("High").shift(-i).over("Symbol").alias(f"Next_High_{i}"))
print("Next 7 days' high prices calculated")

# Function to read and process each df2 file
def process_df2_file(file_path):
    df2 = pl.read_csv(file_path)
    print(f"Processing {file_path} with {df2.shape[0]} rows and {df2.shape[1]} columns")
    # Convert the date column to datetime format
    df2 = df2.with_column(pl.col("date").str.strptime(pl.Date, "%d-%m-%Y"))
    # Select required columns from df1 for join
    join_columns = ["Date", "Symbol", "Close"] + [f"Next_High_{i}" for i in range(1, 8)]
    # Join df2 with df1 to lookup the Close price and Next High prices
    joined_df = df2.join(df1.select(join_columns),
                         left_on=["symbol", "date"], right_on=["Symbol", "Date"],
                         how="left")
    
    # Calculate the Max 7 Days
    next_high_cols = [f"Next_High_{i}" for i in range(1, 8)]
    joined_df = joined_df.with_column(pl.max(next_high_cols).alias("Max_7_Days"))
    
    # Calculate BTST PROFIT %
    joined_df = joined_df.with_column(
        ((pl.col("Next_High_1") - pl.col("Close")) / pl.col("Close") * 100).alias("BTST_PROFIT_%")
    )
    
    # Calculate 7 Day PROFIT %
    joined_df = joined_df.with_column(
        ((pl.col("Max_7_Days") - pl.col("Close")) / pl.col("Close") * 100).alias("7_Day_PROFIT_%")
    )
    
    # Verify the join results
    print(f"Joined DataFrame has {joined_df.shape[0]} rows and {joined_df.shape[1]} columns")
    return joined_df

# Process each file in df2 folder
for file_name in os.listdir(df2_folder):
    if file_name.endswith('.csv'):
        file_path = os.path.join(df2_folder, file_name)
        result = process_df2_file(file_path)
        # Save the result to a new file in the final data folder
        output_file_path = os.path.join(final_data_folder, f"{os.path.splitext(file_name)[0]}_final_data.csv")
        result.write_csv(output_file_path)
        print(f"Saved result to {output_file_path}")

print("Processing complete. Files saved in 'Final Data' folder.")
```
